```python
import pandas

df = pandas.read_excel('test1.xlsx')
df
```

```python
# 모든 학생의 영어점수에 점수 +5
df['영어점수'] = df['영어점수'] + 5
df
```

```python
# 평균 점수 수정
df['평균'] = (df['국어점수'] + df['영어점수']) / 2
df
```

```python
# 컬럼 추가하기
df['결과'] = 'Fail'
df
```

```python
# 특정 조건에 맞는 데이터만 수정
# 국어, 영어의 합이 170점 이상이라는 조건
df.loc[df['국어점수'] + df['영어점수'] >= 170, '결과'] = 'Pass'
df
```

```python
# 컬럼 삭제하기
#df.drop(columns={'결과'}) # 결과 컬럼 삭제
df.drop(columns={'평균', '결과'}, inplace=True) #평균, 결과 컬럼 삭제
```

```python
# 학번에서 '번'이라는 글자 제거
# 힌트 : 문자열은 리스트다. int()안에 문자열을 넣으면 숫자로 나온다.

# DataFrame에서 문자열 데이터를 처리할 때, 컬럼 데이터에 .str이라는 키워드를 붙여서 사용해야함
# df['학번'] = int(df['학번'][:-1]) 오류
#df['학번'] = df['학번'].str.replace('번', '')
df['학번'] = df['학번'].str[:-1]
df
```
